```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk

lemmatizer = WordNetLemmatizer()

# --- Función Auxiliar para mapear Tgas de POS ---
def get_wordnet_pos(nltk_tag):
    """
    Traduce un tag de NLTK (como 'VB') a un tag de WordNet (como 'v').
    """
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        # Por defecto, si no es uno de esos, lo trata como sustantivo
        return wordnet.NOUN

def lemmatize_token_list(token_list):
    """
    Aplica lematización sensible al POS a una lista de tokens.
    """
    # Obtener los tags de POS para cada token (ej. [('managing', 'VBG'), ('projects', 'NNS')])
    pos_tags = nltk.pos_tag(token_list)
    
    # Mapear los tags y lematizar
    lemmatized_tokens = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag)) 
        for word, tag in pos_tags
    ]
    
    return lemmatized_tokens

print("--- Aplicando Lematización (con POS tagging) a 'resume' ---")

try:
    ejemplo_antes = df['resume'].iloc[0][:15]
    
    df['resume'] = df['resume'].apply(lemmatize_token_list)
    
    print("Lematización completada.")
    
    print("\n--- Vista previa de la lematización (Primer currículum) ---")
    
    print("\nANTES (primeros 15 tokens):")
    print(ejemplo_antes)
    
    print("\nDESPUÉS (primeros 15 tokens):")
    print(df['resume'].iloc[0][:15])
    
    print(f"\nEl tipo de dato sigue siendo: {type(df['resume'].iloc[0])}")

except Exception as e:
    print(f"Ocurrió un error inesperado durante la lematización: {e}")
```

## Target

```python
print("--- Creando Target Ponderado ---")

try:
    # Calcular los umbrales para "alto"
    verbs_threshold = df['action_verbs_count'].quantile(0.60)
    skills_threshold = df['soft_skills_count'].quantile(0.75)
    
    print(f"Umbral 'alto' para Action Verbs: {verbs_threshold}")
    print(f"Umbral 'alto' para Soft Skills: {skills_threshold}")

    def get_candidate_score(row):
        score = 0
        
        # +2 puntos por la prueba de idioma (alta prioridad)
        if row['language_test'] == 1:
            score += 2
            
        # +2 puntos por la prueba técnica (alta prioridad)
        if row['technical_test'] == 1:
            score += 3
            
        # +1 punto por conteo alto de verbos
        if row['action_verbs_count'] > verbs_threshold:
            score += 1
            
        # +1 punto por conteo alto de habilidades blandas
        if row['soft_skills_count'] > skills_threshold:
            score += 1
            
        return score

    print("\nCalculando puntuación para cada candidato...")
    df['candidate_score'] = df.apply(get_candidate_score, axis=1)

    # Definir el 'target' basado en un umbral de puntuación
    # (Usaremos >= 4 como un buen punto de corte para "pasar")
    score_threshold = 5
    df['target'] = (df['candidate_score'] >= score_threshold).astype(int)
    
    print(f"Target creado. Candidatos con score >= {score_threshold} son '1' (Pasa).")

    print("\n--- Distribución del Target (Pasa/No Pasa) ---")
    print(df['target'].value_counts(normalize=True))
    
    print("\n--- Distribución de Puntuaciones ---")
    print(df['candidate_score'].value_counts().sort_index())

    print("\n--- Vista previa con score y target ---")
    print(df[['language_test', 'technical_test', 'action_verbs_count', 'soft_skills_count', 'candidate_score', 'target']].head())

except KeyError as e:
    print(f"\nError: Falta la columna {e}.")
    print("Asegúrate de que el DataFrame 'df' tenga todas las columnas necesarias.")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['No Pasa (0)', 'Pasa (1)'], 
                yticklabels=['No Pasa (0)', 'Pasa (1)'])
    plt.title(f'Matriz de Confusión - {model_name}')
    plt.ylabel('Etiqueta Real')
    plt.xlabel('Etiqueta Predicha')
    plt.show()

try:
    
    numeric_features = ['language_test', 'technical_test', 'action_verbs_count', 'soft_skills_count']
    categorical_features = ['category']
    text_feature = 'resume_clean'

    text_transformer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
    categorical_transformer = OneHotEncoder(handle_unknown='ignore')
    numeric_transformer = StandardScaler()

    preprocessor = ColumnTransformer(
        transformers=[
            ('text', text_transformer, text_feature),
            ('categorical', categorical_transformer, categorical_features),
            ('numeric', numeric_transformer, numeric_features)
        ],
        remainder='drop'
    )
    print("Preprocesador (ColumnTransformer) definido.")

    print("\n--- Separando X e y ---")
    
    y = df['target']
    
    # Quitamos 'target' y 'candidate_score' (que se usó para crear el target) de nuestras características para evitar data leakage.
    X = df.drop(columns=['target', 'candidate_score'])
    
    print(f"Forma de X (características): {X.shape}")
    print(f"Forma de y (target): {y.shape}")

    print("\n--- Dividiendo en Train/Test (80/20) ---")
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, 
        test_size=0.20, 
        random_state=42, 
        stratify=y 
    )
    
    print("Aplicando preprocesador (fit_transform en train, transform en test)...")
    
    X_train_processed = preprocessor.fit_transform(X_train)
    
    X_test_processed = preprocessor.transform(X_test)
    
    print("¡Datos listos para entrenar!")
    print(f"Forma de X_train_processed: {X_train_processed.shape}")
    print(f"Forma de X_test_processed: {X_test_processed.shape}")


    # --- Modelo 1: Regresión Logística ---
    print("\n" + "="*40)
    print(" 🚀 Modelo 1: Regresión Logística")
    print("="*40)
    
    lr_model = LogisticRegression(max_iter=1000, random_state=42)
    lr_model.fit(X_train_processed, y_train)
    y_pred_lr = lr_model.predict(X_test_processed)
    
    print(f"Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
    print("\nReporte de Clasificación (LR):")
    print(classification_report(y_test, y_pred_lr))
    plot_confusion_matrix(y_test, y_pred_lr, "Regresión Logística")

    # --- Modelo 2: Random Forest ---
    print("\n" + "="*40)
    print(" 🌳 Modelo 2: Random Forest")
    print("="*40)
    
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    rf_model.fit(X_train_processed, y_train)
    y_pred_rf = rf_model.predict(X_test_processed)
    
    print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
    print("\nReporte de Clasificación (RF):")
    print(classification_report(y_test, y_pred_rf))
    plot_confusion_matrix(y_test, y_pred_rf, "Random Forest")

    # --- Modelo 3: XGBoost ---
    print("\n" + "="*40)
    print(" 🤖 Modelo 3: XGBoost")
    print("="*40)
    
    xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
    xgb_model.fit(X_train_processed, y_train)
    y_pred_xgb = xgb_model.predict(X_test_processed)
    
    print(f"Accuracy: {accuracy_score(y_test, y_pred_xgb):.4f}")
    print("\nReporte de Clasificación (XGB):")
    print(classification_report(y_test, y_pred_xgb))
    plot_confusion_matrix(y_test, y_pred_xgb, "XGBoost")

except NameError as e:
    print(f"\nError: Parece que la variable 'df' no está cargada o falta una columna.")
    print(f"Detalle: {e}")
except Exception as e:
    print(f"\nOcurrió un error inesperado: {e}")
```

```python
lemmatizer = WordNetLemmatizer()
stop_words_set = set(stopwords.words('english'))

def get_wordnet_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def clean_and_process_resume(raw_text):
    text = str(raw_text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    
    tokens = text.split(' ')
    tokens = [token for token in tokens if token]
    
    tokens = [token for token in tokens if token not in stop_words_set]
    
    # Lematización con POS-tagging
    pos_tags = nltk.pos_tag(tokens)
    lemmatized_tokens = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag)) 
        for word, tag in pos_tags
    ]
    
    return ' '.join(lemmatized_tokens)
```

```python
def predecir_candidato(raw_text, category, lang_test, tech_test, preprocessor, model):
    """
    Clasifica un nuevo candidato basándose en sus datos de entrada.
    
    Args:
    - raw_text (str): El texto extraído del PDF.
    - category (str): La categoría del puesto (ej. "data science").
    - lang_test (int): 0 o 1.
    - tech_test (int): 0 o 1.
    - preprocessor: El ColumnTransformer ya entrenado (con .fit()).
    - model: El modelo de ML ya entrenado (ej. rf_model).
    """
    
    print("--- Iniciando nueva predicción ---")
    
    print("Calculando conteo de palabras clave...")
    action_verbs = count_words(raw_text, ACTION_VERBS)
    soft_skills = count_words(raw_text, SOFT_SKILLS)
    
    print("Limpiando y procesando texto del currículum (NLP)...")
    resume_clean_text = clean_and_process_resume(raw_text)
    
    data_dict = {
        'resume_clean': [resume_clean_text],
        'category': [category.lower()], 
        'language_test': [lang_test],
        'technical_test': [tech_test],
        'action_verbs_count': [action_verbs],
        'soft_skills_count': [soft_skills]
    }
    
    new_data_df = pd.DataFrame(data_dict)
    print("DataFrame de entrada creado:")
    print(new_data_df.head())
    
    # --- Aplicar el Preprocesador (SOLO .transform()) ---
    print("Aplicando preprocesador (TF-IDF, OneHot, Scaler)...")
    X_processed = preprocessor.transform(new_data_df)
    
    # --- Predicción ---
    print("Realizando predicción...")
    prediction = model.predict(X_processed)
    probabilities = model.predict_proba(X_processed)
    
    # --- Resultado ---
    resultado_clase = "Pasa (1)" if prediction[0] == 1 else "No Pasa (0)"
    confianza = probabilities[0][prediction[0]] * 100
    
    print("--- ¡Predicción Completa! ---")
    return {
        'clase_predicha': resultado_clase,
        'confianza': f"{confianza:.2f}%"
    }
```

```python
import fitz

def extraer_texto_de_pdf(ruta_pdf):
    """
    Extrae todo el texto de un archivo PDF usando PyMuPDF (fitz).
    """
    try:
        documento = fitz.open(ruta_pdf)
    except Exception as e:
        print(f"Error al abrir el archivo {ruta_pdf}: {e}")
        return None

    texto_completo = ""

    for num_pagina in range(documento.page_count):
        pagina = documento.load_page(num_pagina)
        
        texto_pagina = pagina.get_text()
        
        texto_completo += texto_pagina + "\n" # Añadimos un salto de línea por página

    # Cerrar el documento
    documento.close()
    
    return texto_completo

ruta_archivo = "./data/cv-isaijesus.pdf"  
texto_cv = extraer_texto_de_pdf(ruta_archivo)

if texto_cv:
    print("--- Texto Extraído del PDF ---")
    print(texto_cv)
    print("\n------------------------------")
    print(f"Total de caracteres extraídos: {len(texto_cv)}")
```
